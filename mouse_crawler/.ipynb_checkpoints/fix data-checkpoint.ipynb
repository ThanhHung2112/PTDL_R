```python
url = "https://raw.githubusercontent.com/ThanhHung2112/PTDL_R/main/mouse_crawler/mouse_tiki_data.csv"
#df = pd.read_csv(url)
df = pd.read_csv("mouse_tiki_data.csv")
```

```python
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
df
```

```python
def remove_object(c_name,de_str, n=0):
    for i in df[str(c_name)]:
        df[str(c_name)][n] = i.strip(str(de_str))
        n =+ 1
    return
```

```python
remove_object('Price',' ₫')
```

```python
remove_object('Discount','-')
```

```python
n = 0
for i in df["Sold"]:
    df["Sold"][n] = i.strip("Đã bán ")
    n+=1
```

```python
n = 0
for i in df["Price"]:
    df["Price"][n] = i.strip(" ₫")
    n+=1

    
```

```python
n = 0
for i in df["Price"]:
    if ' ₫' in i:
        df["Price"][n] = i[:i.index(' ₫')]
        df['Product_price'][n] = i[i.index(' ₫')+3:i.index('₫-')]
        df['Discount'][n] = i[i.index('₫-')+2:]
        print(i.index(' ₫'))
    n =+ 1
```
