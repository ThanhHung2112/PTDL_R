```python
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
def remove_object(c_name,de_str, n=0):
    for i in df[str(c_name)]:
        df[str(c_name)][n] = i.strip(str(de_str))
        n += 1
    return
```

```python
n = 0
for i in df['Warranty']:
    i = str(i)
    if ' Tháng' in i : df['Warranty'][n] = float(i[:i.index(' Tháng')])
    elif ' Năm' in i: df['Warranty'][n] = float(i[:i.index(' Năm')])*12
    else: df['Warranty'][n] = float(i)
    n += 1
        
```

```python
def fixnon(data, n = 0):
    for i in data:
        if i == 'Non' : data[n] = 0
        elif i =='No' : data[n] = 0
        else: data[n] = float(i)
        n += 1
    return
```

```python
n = 0
for i in df['Rate']:
    if i == 'Non' :
        df['Rate'][n] = 0
    elif i =='No' :
        df['Rate'][n] = 0
    else:
        df['Rate'][n] = float(i)
    n = n + 1
```
